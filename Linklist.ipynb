```python
class Node:
    def __init__(self, data = None):
        self.data = data
        self.next = None

class Sll:
    def __init__(self):
        self.start = None
    
    def displaylist(self):
        disp = self.start
        while disp is not None:
            print(disp.data)
            disp = disp.next
            
    def AddAtStart(self, data):
        #Create a new node
        TempNode = Node(data)
        #Update the next field of Tempnode to the pesent start node
        TempNode.next = self.start
        #Update the start of the LL to point the Tempnode as the new First node of LL
        self.start =  TempNode
        
    def AddAtEnd(self, data):
        #Create a new node
        TempNode = Node(data)
        #Check whether the LL is empty and so new node will be the first node
        if self.start is None:
            self.start = TempNode
            return
        # Lower code will work if the LL is existing
        #Copy the ref of first node of LL
        lastn = self.start
        #Move the ref to the current last of the LL
        while lastn.next is not None:
            lastn = lastn.next
        #Link the new node as the mew last node of LL
        lastn.next = TempNode
    
    def AddInBetween(self,insertafter, data):
        if insertafter is None:
            print("The node after which the insertion is to be done does not exist")
            return
               
        #Create a new node
        TempNode = Node(data)
        #Assign the next field of current node to next field of new node
        TempNode.next = insertafter.next
        # Assign new node as next of current node
        insertafter.next = TempNode
       
    def DelFrom(self, Key):
        
        #Checks whether deleting from an empty lsit
        if self.start is None:
            print("Nothing to delete as the list is empty")
            return
            
        Temp = self.start
        if(Temp.data == Key): #Means you are deleting the first node
            self.start = Temp.next #LL now starts from second node
            Temp = None 
            return
        
        # Deletion is to be done on any node other than the first node
        prev= Temp
        while(Temp is not None):
            if(Temp.data == Key):
                break
            prev= Temp
            Temp=Temp.next
        if (Temp == None):
                return
            
        prev.next = Temp.next
        Temp = None
        
        
        
        
        
        
        
        
        

# Created an emptylist        
linked_list = Sll()

# Created the first node and assigned its address to the start pointer of the linked list
linked_list.start= Node("Feb")

#Created independent nodes
n1 = Node("April")
n2 = Node("May")
n3 = Node("June")


#The second node is added in the LL
linked_list.start.next = n1

#Add the third node
n1.next = n2

#Link the fourth node
n2.next = n3

#Add at Start
linked_list.AddAtStart("Jan")

#Add at End

linked_list.AddAtEnd("July")


#Add in between

linked_list.AddInBetween(linked_list.start.next, "March")


linked_list.DelFrom("April")
linked_list.displaylist()

```
